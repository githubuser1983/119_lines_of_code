```python
import numpy as np

def writePitches(fn,inds,tempo=82,instrument=[0,0],add21=True,start_at= [0,0],durationsInQuarterNotes=False):
    from MidiFile import MIDIFile

    track    = 0
    channel  = 0
    time     = 0   # In beats
    duration = 1   # In beats # In BPM
    volume   = 116 # 0-127, as per the MIDI standard

    ni = len(inds)
    MyMIDI = MIDIFile(ni,adjust_origin=False) # One track, defaults to format 1 (tempo track
                     # automatically created)
    MyMIDI.addTempo(track,time, tempo)


    for k in range(ni):
        MyMIDI.addProgramChange(k,k,0,instrument[k])


    times = start_at
    for k in range(len(inds)):
        channel = k
        track = k
        for i in range(len(inds[k])):
            pitch,duration,volume,isPause,tempo = inds[k][i]
            #print(pitch,duration,volume,isPause)
            track = k
            channel = k
            if not durationsInQuarterNotes:
                duration = 4*duration#*maxDurations[k] #findNearestDuration(duration*12*4)            
            #print(k,pitch,times[k],duration,100)
            if not isPause: #rest
                #print(volumes[i])
                # because of median:
                pitch = int(np.floor(pitch))
                if add21:
                    pitch += 21
                #print(pitch,times[k],duration,volume,isPause)    
                MyMIDI.addTempo(track,times[k], tempo)
                MyMIDI.addNote(track, channel, int(pitch), float(times[k]) , float(duration), int(volume))
                times[k] += duration*1.0  
            else:
                times[k] += duration*1.0
       
    with open(fn, "wb") as output_file:
        MyMIDI.writeFile(output_file)
    print("written")  


   
def HH(start,end):
    H = Graph(loops=False)
    for a in range(start,end+1):
        H.add_vertex(a)
        for b in range(start,end+1):
            if abs(a-b) in [3,4]:
                H.add_edge(a,b,abs(a-b))
    return H            

def jacc(a,b):
    return len(set(a).intersection(set(b)))/len(set(a).union(set(b)))

def GG(start,end):
    H = HH(start,end)
    cb = H.cycle_basis()
    G = Graph(loops = False)
    for c1 in cb:
        G.add_vertex(tuple(c1))
        for c2 in cb:
            if jacc(c1,c2)>0 and c1!=c2:
                G.add_edge(tuple(c1),tuple(c2))
    return G            

def volumeByPitch(p,st,en):
    return en-abs(p-(en+st)//2)

iinds = []

ss = []
st = 32
en = 64+st
G = GG(st,en)
G.plot().show()
hamiltonian_path = list(G.hamiltonian_path())
#mcb = (G.cycle_basis())
    #dat = data[i]

    #start = i%N+1
iinds = [[] for k in range(20)]    
for j in range(2*len(hamiltonian_path)):
    sl = hamiltonian_path[j//2]
    c = 0
    sch = sorted([(jacc(x,sl),x) for x in G[sl]])
    ch = []
    #ch.append(sch[len(sch)-4])
    #ch.append(sch[len(sch)-3])
    ch.append(sch[len(sch)-2])
    ch.append(sch[len(sch)-1])
    print("ch = ",ch)
    for ii in ch:
        #notes = []
        for t in ii[1]:
            #print(sl)
            print("t = ",t,ii[1])
            tempo = 80
            isPause = False #(stlen(ii)%7<2) #(t%3==0)
            pitch = t
            volume = volumeByPitch(t,st,en)
            duration = np.log(t)/sum([np.log(x) for x in ii[1]])
            note = (pitch,duration,volume,isPause,tempo)
            print("note=",note)
            iinds[c].append(note)
        c+=1          


fn = "./midi/experiment_"+str(st)+"-"+str(en)+".mid"    
writePitches(fn,iinds,tempo=80,instrument=len(iinds)*[81],add21=False,start_at= len(iinds)*[0],durationsInQuarterNotes=False)    
```
